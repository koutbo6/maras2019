# What is a Webpage?

- **Webpage**: A hyperlinked document (i.e. text file) written in HTML/CSS/Javascript that can be viewed in a web browser.
- **Website**: A collection of webpages that are hyperlinked together.
- **Webserver**: The software that serves the webpage files to the browsers over the internet. The browser talks to the webserver using HTTP.

# Webpage Vs. Webapp?
- Webpages are HTML based, and are static.
- We use languages such as Python/PHP/Ruby ..etc to make them dynamic (i.e., change the content depending on who is using it)
- Webservers can only serve static HTML files, so when they get a request for a dynamic page it forwards it to the application server
- Application server generates the HTML on the fly and sends it to the web server to serve to the browser, creating the illusion that HTML is dynamic
- The code we write using languages like Python is the code to generate HTML that users will see

# Server-Side Vs. Client-Side Web Applications?
- With Server-side applications, most of the work done to produce dynamic HTML is done in the application server and served by the webserver.
- Everytime an action is performed the server-side application will have to reload on screen.
- With advances in browser technology, it is now possible to let the browser update the page with new information from the webserver/appserver and make it seem more dynamic
- With new updates from the webserver, the client-side application loads only the portion of the page that changed, and does not reload the whole page

# Web Frameworks
- Group of tools, functions, software, and guideline, if followed, will simplify the development of webapps and provide some guarantees to the developer about the application
- Developers must organize their files in a certain way, or name them a certain way.
- Developers gain by reducing the effort to develop web applications significantly.
- Frameworks differ in the way they standardize development and the benefits they bring to the developer.

# Request/Response Cycle in All Web Applications
1. Browser sends and HTTP request to webserver
2. Webserver forwards the HTTP request to the Application Server
3. Application server gets the user data in the request and performs the required business process
4. The result is prepared as HTML data, included in the HTTP response, and sent back to the webserver
5. The webserver forwards the HTTP response to the browser to display

When you do web application development, you work on steps 3 & 4

# Micro Vs. Fullstack Web Framework
- Micro frameworks are very simplistic, they allow users to handle HTTP requests from webserver and create HTTP responses.
    + Advantage: Simple to learn and deploy, but not feature rich.
    + Example: Flask and Bottle
    
- Fullstack frameworks offer complete solutions for a data driven web application, including how databases are handled, and how HTML responses are constructed, in addition to a whole host of features.
    + Advantage: Offers numerous features, great for rapid development, but needs time to learn.
    + Example: Django

# Our Database Application
- Use Sqlite database
- Single user database, no user credential needed.
- Database is a single file
- Get data [euro soccer data set from kaggle](https://www.kaggle.com/hugomathien/soccer)


````python
# step one: load sqlite library (needed once)
import sqlite3

# step two: establish db conenction (we keep reusing the connection, needed once)
conn = sqlite3.connect("database.sqlite")

#step three: make query using the connection
cur = conn.execute("SELECT * FROM player")

#step four: fetch the result (list of tuples)
result = cur.fetchall()
````

````python
# the results is a list of tuples
result
````

# Schema
- See [euro soccer data set from kaggle](https://www.kaggle.com/hugomathien/soccer) to get the schema
- You can also use sqlite program to access the database using the command line
- Do a search on the sqlite command and see how you can find the table names and columns in a database

````python
# let's calculate the average weight for all players
# its the last item in each record (tuple)

player_count = len(result)
weight_sum = 0
for record in result:
    weight_sum += record[-1]
print("average is ", weight_sum/player_count)
````

# Web Application Can Produce Graphs
- You can include these graphs in HTML templates

```python
@web_site.route('/fig/')
def fig():
  conn = sqlite3.connect("database.sqlite")
  df = pd.read_sql("SELECT * FROM Player_Attributes limit 100", conn)
  g = df["preferred_foot"].value_counts().plot(kind="pie")
  fig = g.figure
  img = BytesIO()
  fig.savefig(img)
  img.seek(0)
  return send_file(img, mimetype='image/png')
```
